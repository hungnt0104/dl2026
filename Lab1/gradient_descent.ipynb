```python
def f(x):
    return x**2

def f_prime(x):
    return 2*x

def gradient_descent(x, lr, thres):
    change = float('inf')
    while (change > thres):
        x_new = x - f_prime(x) * lr
        change = f(x) - f(x_new)
        x = x_new
        print(f"f(x): {f(x_new)}, x: {x_new}")

for lr in range(1, 10):
    print("lr = ", lr/10)
    gradient_descent(10, lr/10, 0.01)
    print("================================")
```
