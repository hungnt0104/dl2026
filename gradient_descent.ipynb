```python
def f(x):
    return x**2

def f_prime(x):
    return 2*x

def gradient_descent(x, lr, num_iter):
    for i in range(num_iter):
        x = x - f_prime(x) * lr
        print(f"f(x): {f(x)}, x: {x}")

for lr in range(1, 10):
    print("lr = ", lr/10)
    gradient_descent(10, lr/10, 10)
    print("================================")
```
